```python
# testing all endpoint in quiver quant for <500> response

import requests 
import pandas as pd

base_url = "https://api.quiverquant.com/beta/"
ticker = "MSFT" #for collecting historical data on a single ticker

category = {
    "both":[
        "congresstrading",
        "senatetrading",
        "housetrading",
        "govcontractsall",
        "govcontracts",
        "lobbying",
        "offexchange",
        "wikipedia",
        "wallstreetbets",
        "spacs",
        "twitter",
        "politicalbeta"
    ],
    
    "live_only": [
        "appratings",
        "sec13f",
        "sec13fchanges",
        "etfholdings",
        "insiders"
    ]
}

cat_list = []
endpoint_list = []
live_hist_list = []
http_list = []

for cat in category["both"]:

    cat_list.append(cat)
    live_endpoint = base_url + f"live/{cat}"
    endpoint_list.append(live_endpoint)
    live_hist_list.append("LIVE")
    http_list.append(requests.get(live_endpoint, headers=headers))
    
    cat_list.append(cat)
    hist_endpoint = base_url + f"historical/{cat}/{ticker}"
    endpoint_list.append(hist_endpoint)
    live_hist_list.append("HISTORICAL")
    http_list.append(requests.get(hist_endpoint, headers=headers))
    
for cat in category["live_only"]:
    cat_list.append(cat)
    live_endpoint = base_url + f"live/{cat}"
    endpoint_list.append(live_endpoint)
    live_hist_list.append("LIVE")
    http_list.append(requests.get(live_endpoint, headers=headers))
    
    
pd_lists = [cat_list, live_hist_list, http_list, endpoint_list]
columns = ["Category", "Live/Hist", "HTTP_Response", "Endpoint"]
response_df = pd.DataFrame(pd_lists).transpose()
response_df.columns = columns
    
print(response_df)
```
